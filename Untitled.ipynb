```python
import requests
import json
import time

# ---------- CONFIG ----------
BASE_URL = "https://world.openfoodfacts.org/api/v2/search"
PAGE_SIZE = 100
MAX_PAGES = 120           # 100 * 120 = 12,000 products (approx)
OUTPUT_FILE = "chocolates_products.json"
# ----------------------------


def extract_product_fields(product):
    """Pick only the required fields from each product."""
    return {
        "product_code": product.get("code"),
        "product_name": product.get("product_name"),
        "brand": product.get("brands"),
        "nutriments": product.get("nutriments", {})
    }


def main():
    all_products = []

    for page in range(1, MAX_PAGES + 1):
        params = {
            "categories": "chocolates",
            "fields": "code,product_name,brands,nutriments",
            "page_size": PAGE_SIZE,
            "page": page
        }

        # ---------- RETRY LOGIC ----------
        success = False
        for attempt in range(3):
            try:
                response = requests.get(BASE_URL, params=params, timeout=60)
                response.raise_for_status()
                success = True
                break  #  success → exit retry loop
            except requests.exceptions.ReadTimeout:
                print(f" Timeout on page {page} (try {attempt+1}/3)... retrying in 5 sec")
                time.sleep(5)
            except requests.exceptions.RequestException as e:
                print(f" Error fetching page {page}: {e}")
                break

        if not success:
            print(f" Skipping page {page} after 3 failed attempts.")
            continue  #  this is inside the for-loop, now valid
        # ---------------------------------

        data = response.json()
        products = data.get("products", [])
        if not products:
            print(f"No more products found at page {page}. Stopping.")
            break

        for p in products:
            all_products.append(extract_product_fields(p))

        print(f" Page {page} fetched. Total so far: {len(all_products)}")

        #  Save every 10 pages as a checkpoint
        if page % 10 == 0:
            with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
                json.dump(all_products, f, ensure_ascii=False, indent=2)
            print(f" Saved checkpoint at page {page}")

        # Be nice to API (avoid hitting it too fast)
        time.sleep(1)

    # ---------- FINAL SAVE ----------
    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        json.dump(all_products, f, ensure_ascii=False, indent=2)
    print(f"\n Done! Saved {len(all_products)} products to {OUTPUT_FILE}")


# ---------- RUN ----------
if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import json

# Load the JSON file
with open("chocolates_products.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Convert nested JSON to DataFrame
df = pd.json_normalize(data)

# Show basic info
print(" Data Loaded Successfully!")
print("Shape:", df.shape)
df.head(3)
```

```python
# Keep only useful columns
keep_cols = [
    "product_code",
    "product_name",
    "brand",
    "nutriments.energy-kcal_value",
    "nutriments.energy-kj_value",
    "nutriments.carbohydrates_value",
    "nutriments.sugars_value",
    "nutriments.fat_value",
    "nutriments.saturated-fat_value",
    "nutriments.proteins_value",
    "nutriments.fiber_value",
    "nutriments.salt_value",
    "nutriments.sodium_value",
    "nutriments.nova-group",
    "nutriments.nutrition-score-fr",
    "nutriments.fruits-vegetables-nuts-estimate-from-ingredients_100g"
]

df = df[keep_cols].copy()

# Rename columns for simplicity
df.columns = [
    "product_code", "product_name", "brand",
    "energy_kcal", "energy_kj", "carbohydrates", "sugars",
    "fat", "saturated_fat", "proteins", "fiber", "salt", "sodium",
    "nova_group", "nutrition_score_fr", "fruits_veg_nuts"
]

df.head(3)
```

```python
# Replace empty strings with NaN
df = df.replace("", pd.NA)

# Convert numeric columns
num_cols = [
    "energy_kcal", "energy_kj", "carbohydrates", "sugars", "fat",
    "saturated_fat", "proteins", "fiber", "salt", "sodium",
    "nova_group", "nutrition_score_fr", "fruits_veg_nuts"
]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")

# Check for missing values
df.isna().sum()
```

```python
# 1️ Sugar-to-Carb Ratio
df["sugar_to_carb_ratio"] = (df["sugars"] / df["carbohydrates"]).round(2)

# 2️ Calorie Category
def calorie_category(kcal):
    if pd.isna(kcal):
        return "Unknown"
    elif kcal < 200:
        return "Low Calorie"
    elif kcal < 400:
        return "Moderate Calorie"
    else:
        return "High Calorie"

df["calorie_category"] = df["energy_kcal"].apply(calorie_category)

# 3️ Sugar Category
def sugar_category(sugar):
    if pd.isna(sugar):
        return "Unknown"
    elif sugar < 10:
        return "Low Sugar"
    elif sugar < 25:
        return "Moderate Sugar"
    else:
        return "High Sugar"

df["sugar_category"] = df["sugars"].apply(sugar_category)

# 4️ Ultra-Processed Flag
df["is_ultra_processed"] = df["nova_group"].apply(lambda x: "Yes" if x == 4 else "No")

df.head(5)
```

```python
import pandas as pd
from sqlalchemy import create_engine
import json

# ---------- CONFIG ----------
DB_USER = "root"
DB_PASS = "root"    
DB_HOST = "localhost"
DB_NAME = "choco_crunch"
JSON_FILE = "chocolates_products.json"
# ----------------------------

print(" Reading JSON data...")
df = pd.read_json(JSON_FILE)

#  Flatten the 'nutriments' field if present
if 'nutriments' in df.columns:
    print(" Flattening nested 'nutriments' field...")
    nutriments_df = pd.json_normalize(df['nutriments'])
    df = pd.concat([df.drop(columns=['nutriments']), nutriments_df], axis=1)

#  Convert dict/list cells to JSON strings
for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, (dict, list))).any():
        df[col] = df[col].apply(lambda x: json.dumps(x) if isinstance(x, (dict, list)) else x)

#  Clean inf / -inf / NaN
df = df.replace([float("inf"), float("-inf")], None)

#  Ensure column names are valid for MySQL
max_len = 60
rename_map = {}
for i, col in enumerate(df.columns):
    clean_col = col.strip().replace(" ", "_").replace("-", "_")
    if len(clean_col) > max_len:
        new_name = f"col_{i}"
        rename_map[col] = new_name
        print(f"⚙️ Truncated long column name → {new_name}")
    elif clean_col != col:
        rename_map[col] = clean_col
df = df.rename(columns=rename_map)

#  Drop duplicate columns
df = df.loc[:, ~df.columns.duplicated()]

#  Keep only selected important columns
keep_cols = [
    "code", "product_name", "brands",
    "energy_100g", "fat_100g", "saturated_fat_100g",
    "carbohydrates_100g", "sugars_100g", "fiber_100g",
    "proteins_100g", "salt_100g", "sodium_100g"
]
df = df[[c for c in keep_cols if c in df.columns]]

print(" Columns selected for upload:", df.columns.tolist())

#  Connect and upload to MySQL
print(" Connecting to MySQL...")
engine = create_engine(f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}/{DB_NAME}")

print(" Uploading data to MySQL (please wait)...")
try:
    df.to_sql("chocolates", con=engine, if_exists="replace", index=False)
    print(f" Successfully loaded {len(df)} rows into MySQL table 'chocolates'!")
except Exception as e:
    print(" Error while uploading to MySQL:", e)
```

```python
import mysql.connector

# ✅ Connect to your MySQL database
conn = mysql.connector.connect(
    host="localhost",
    user="root",              
    password="root", 
    database="choco_crunch"    
)
cursor = conn.cursor()

# --- Create product_info table ---
cursor.execute("""
CREATE TABLE IF NOT EXISTS product_info (
  product_code VARCHAR(255) PRIMARY KEY,
  product_name TEXT,
  brand TEXT
);
""")

# --- Create nutrient_info table ---
cursor.execute("""
CREATE TABLE IF NOT EXISTS nutrient_info (
  product_code VARCHAR(255) PRIMARY KEY,
  energy_kcal_value FLOAT,
  energy_kj_value FLOAT,
  carbohydrates_value FLOAT,
  sugars_value FLOAT,
  fat_value FLOAT,
  saturated_fat_value FLOAT,
  proteins_value FLOAT,
  fiber_value FLOAT,
  salt_value FLOAT,
  sodium_value FLOAT,
  fruits_vegetables_nuts_estimate_from_ingredients_100g FLOAT,
  nutrition_score_fr INT,
  nova_group INT,
  FOREIGN KEY (product_code) REFERENCES product_info(product_code)
);
""")

# --- Create derived_metrics table ---
cursor.execute("""
CREATE TABLE IF NOT EXISTS derived_metrics (
  product_code VARCHAR(255) PRIMARY KEY,
  sugar_to_carb_ratio FLOAT,
  calorie_category TEXT,
  sugar_category TEXT,
  is_ultra_processed TEXT,
  FOREIGN KEY (product_code) REFERENCES product_info(product_code)
);
""")

# Save and close
conn.commit()
conn.close()

print(" All 3 tables created successfully in MySQL database (choco_crunch)!")
```

```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine, text

# -------------------------------
#  Load dataset
# -------------------------------
file_path = r"C:\Users\win-10\Desktop\Choco_Crunch Analytics\cleaned_chocolates.csv"
df = pd.read_csv(file_path)
print(" Data loaded successfully!")

# -------------------------------
#  Connect to MySQL
# -------------------------------
username = "root"
password = "root"
host = "localhost"
database = "choco_crunch"

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}/{database}")

# -------------------------------
#  Drop existing tables safely
# -------------------------------
with engine.connect() as conn:
    conn.execute(text("SET FOREIGN_KEY_CHECKS=0;"))
    conn.execute(text("DROP TABLE IF EXISTS derived_metrics;"))
    conn.execute(text("DROP TABLE IF EXISTS nutrient_info;"))
    conn.execute(text("DROP TABLE IF EXISTS product_info;"))
    conn.execute(text("SET FOREIGN_KEY_CHECKS=1;"))
    print(" Old tables dropped.")

# -------------------------------
#  Create fresh tables
# -------------------------------
create_table_sql = """
CREATE TABLE product_info (
    product_code BIGINT PRIMARY KEY,
    product_name TEXT,
    brand TEXT
);

CREATE TABLE nutrient_info (
    id INT AUTO_INCREMENT PRIMARY KEY,
    product_code BIGINT,
    energy_kcal FLOAT,
    energy_kj FLOAT,
    carbohydrates FLOAT,
    sugars FLOAT,
    fat FLOAT,
    saturated_fat FLOAT,
    proteins FLOAT,
    fiber FLOAT,
    salt FLOAT,
    sodium FLOAT,
    fruits_veg_nuts FLOAT,
    nutrition_score_fr FLOAT,
    nova_group INT,
    FOREIGN KEY (product_code) REFERENCES product_info(product_code)
);

CREATE TABLE derived_metrics (
    id INT AUTO_INCREMENT PRIMARY KEY,
    product_code BIGINT,
    sugar_to_carb_ratio FLOAT,
    calorie_category VARCHAR(50),
    sugar_category VARCHAR(50),
    is_ultra_processed VARCHAR(10),
    FOREIGN KEY (product_code) REFERENCES product_info(product_code)
);
"""

with engine.connect() as conn:
    for stmt in create_table_sql.strip().split(";"):
        if stmt.strip():
            conn.execute(text(stmt))
    print(" Tables created successfully!")

# -------------------------------
#  Split dataframes
# -------------------------------
product_info = df[['product_code', 'product_name', 'brand']]

nutrient_info = df[['product_code', 'energy_kcal', 'energy_kj', 'carbohydrates',
                    'sugars', 'fat', 'saturated_fat', 'proteins', 'fiber',
                    'salt', 'sodium', 'fruits_veg_nuts', 'nutrition_score_fr', 'nova_group']]

derived_metrics = df[['product_code', 'sugar_to_carb_ratio', 'calorie_category',
                      'sugar_category', 'is_ultra_processed']]

# -------------------------------
#  Clean invalid values (critical fix)
# -------------------------------
for d in [product_info, nutrient_info, derived_metrics]:
    d.replace([np.inf, -np.inf], np.nan, inplace=True)
    d.fillna(value=np.nan, inplace=True)

# Optional: convert NaN to None for MySQL NULLs
product_info = product_info.where(pd.notnull(product_info), None)
nutrient_info = nutrient_info.where(pd.notnull(nutrient_info), None)
derived_metrics = derived_metrics.where(pd.notnull(derived_metrics), None)

# -------------------------------
#  Insert data (append mode)
# -------------------------------
product_info.to_sql("product_info", con=engine, if_exists="append", index=False)
nutrient_info.to_sql("nutrient_info", con=engine, if_exists="append", index=False)
derived_metrics.to_sql("derived_metrics", con=engine, if_exists="append", index=False)

print(" All data successfully inserted into all 3 tables!")
```

```python
import pandas as pd

# ---------------------------------------------
# LOAD CSV FILES
# ---------------------------------------------
product_info = pd.read_csv("product_info.csv")
nutrient_info = pd.read_csv("nutrient_info.csv")
derived_metrics = pd.read_csv("derived_metrics.csv")

# Merge master dataset
merged = (product_info
          .merge(nutrient_info, on="product_code", how="left")
          .merge(derived_metrics, on="product_code", how="left"))

print(" Data Loaded & Merged")

# ---------------------------------------------
# CREATE A FOLDER TO SAVE RESULTS
# ---------------------------------------------
import os
os.makedirs("query_results", exist_ok=True)

def save(df, name):
    path = f"query_results/{name}.csv"
    df.to_csv(path, index=False)
    print(f"✔ Saved: {path}")

# ---------------------------------------------
#  27 QUERIES (UPDATED WITH CORRECT COLUMNS)
# ---------------------------------------------

# 1. Total products
q1 = merged[['product_code']].nunique()
save(q1, "01_total_products")

# 2. Products by brand
q2 = merged.groupby('brand')['product_code'].nunique().reset_index()
save(q2, "02_products_by_brand")

# 3. Top 10 sugary products
q3 = merged.nlargest(10, 'sugars')[['product_name','brand','sugars']]
save(q3, "03_top10_sugar")

# 4. Top 10 fat products
q4 = merged.nlargest(10, 'fat')[['product_name','brand','fat']]
save(q4, "04_top10_fat")

# 5. Avg nutrition per calorie category
q5 = merged.groupby('calorie_category')[['energy_kcal','sugars','fat']].mean().reset_index()
save(q5, "05_avg_nutrition_calorie_category")

# 6. Products with zero sugar
q6 = merged[merged['sugars'] == 0][['product_name','brand']]
save(q6, "06_zero_sugar")

# 7. Products with sugar > 30g
q7 = merged[merged['sugars'] > 30][['product_name','brand','sugars']]
save(q7, "07_high_sugar")

# 8. Count by calorie category
q8 = merged['calorie_category'].value_counts().reset_index()
save(q8, "08_calorie_category_count")

# 9. NOVA classification
q9 = merged.groupby('nova_group')['product_code'].nunique().reset_index()
save(q9, "09_nova_distribution")

# 10. Ultra processed products
q10 = merged[merged['is_ultra_processed'] == 1][['product_name','brand']]
save(q10, "10_ultra_processed")

# 11. Top 5 brands by product count
q11 = merged['brand'].value_counts().head(5).reset_index()
save(q11, "11_top5_brands")

# 12. Lowest calorie items
q12 = merged.nsmallest(10, 'energy_kcal')[['product_name','brand','energy_kcal']]
save(q12, "12_lowest_calorie")

# 13. Highest protein products
q13 = merged.nlargest(10, 'proteins')[['product_name','brand','proteins']]
save(q13, "13_highest_protein")

# 14. Sugar vs Calories correlation
q14 = merged[['sugars','energy_kcal']].corr()
save(q14, "14_correlation_sugar_calories")

# 15. Fat vs Calories correlation
q15 = merged[['fat','energy_kcal']].corr()
save(q15, "15_correlation_fat_calories")

# 16. Products with missing nutrition
q16 = merged[merged['energy_kcal'].isna()]
save(q16, "16_missing_nutrition")

# 17. Calorie category - max calories
q17 = merged.groupby('calorie_category')['energy_kcal'].max().reset_index()
save(q17, "17_max_calories_calorie_category")

# 18. Calorie category - min calories
q18 = merged.groupby('calorie_category')['energy_kcal'].min().reset_index()
save(q18, "18_min_calories_calorie_category")

# 19. Calorie category - median calories
q19 = merged.groupby('calorie_category')['energy_kcal'].median().reset_index()
save(q19, "19_median_calories_calorie_category")

# 20. Sugar distribution (bins)
q20 = pd.cut(
    merged['sugars'], 
    bins=[0,5,15,30,100], 
    labels=['Very Low','Low','Medium','High']
)
q20 = q20.value_counts().reset_index()
save(q20, "20_sugar_distribution_bins")

# 21. High salt products
q21 = merged[merged['salt'] > 1][['product_name','brand','salt']]
save(q21, "21_high_salt")

# 22. High fiber products
q22 = merged[merged['fiber'] > 5][['product_name','brand','fiber']]
save(q22, "22_high_fiber")

# 23. Avg nutrition of top 10 brands
top_brands = merged['brand'].value_counts().head(10).index
q23 = merged[merged['brand'].isin(top_brands)]\
       .groupby('brand')[['energy_kcal','sugars','fat']].mean().reset_index()
save(q23, "23_avg_nutrition_top_brands")

# 24. Sugar ranking by calorie category
q24 = merged.groupby('calorie_category')['sugars']\
            .mean().sort_values(ascending=False).reset_index()
save(q24, "24_calorie_category_sugar_ranking")

# 25. Fat ranking by calorie category
q25 = merged.groupby('calorie_category')['fat']\
            .mean().sort_values(ascending=False).reset_index()
save(q25, "25_calorie_category_fat_ranking")

q26 = merged[merged['proteins'] > 8][['product_name','brand','proteins']]
save(q26, "26_protein_rich")

# 27. Final master table
save(merged, "27_master_table")

print("\n ALL 27 QUERIES EXECUTED & SAVED!")

```
